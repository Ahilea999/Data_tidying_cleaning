```python
coffee_data_merged2["month"].unique()
```

```python
def month_name_to_number(month_name):
    if month_name == "January":
        return "01"
    elif month_name == "February":
        return "02"
    elif month_name == "March":
        return "03"
    elif month_name == "April":
        return "04"
    elif month_name == "May":
        return "05"
    elif month_name == "June":
        return "06"
    elif month_name == "July":
        return "07"
    elif month_name == "August":
        return "08"
    elif month_name == "September":
        return "09"
    elif month_name == "October":
        return "10"
    elif month_name == "November":
        return "11"
    elif month_name == "December":
        return "12"
    else:
        return "Invalid month name"
```

```python
coffee_data_merged2["month"] = coffee_data_merged2.month.apply(month_name_to_number)
```

```python
coffee_data_merged2.month
```

```python
coffee_data_merged2["grading_date"] = coffee_data_merged2["year"].astype(str) + "-" + coffee_data_merged2["month"].astype(str) + "-" + coffee_data_merged2["day"].astype(str)
```

```python
def country_to_continent(country):
    if country in ["Costa Rica", "United States*", "Guatemala", "Haiti", "Mexico", "Nicaragua", "Honduras", "Panama", "El Salvador"]:
        return "North America"
    elif country in ["Peru", "Ecuador", "Colombia", "Brazil"]:
        return "South America"
    elif country in ["India", "Vietnam", "China", "Taiwan", "Thailand", "Indonesia", "Philippines", "Laos", "Myanmar"]:
        return "Asia"
    elif country in ["Papua New Guinea"]:
        return "Oceania"
    elif country in ["Ethiopia", "Uganda", "Tanzania, United Republic Of", "Kenya", "Burundi", "Rwanda", "Malawi", "Zambia", "Mauritius", "Cote d?Ivoire"]:
        return "Africa"         
```

```python
coffee_data_merged3["continent"] = coffee_data_merged3.country_of_origin.apply(country_to_continent)
```

```python
coffee_data_merged3[["continent", "color"]].value_counts()
```
